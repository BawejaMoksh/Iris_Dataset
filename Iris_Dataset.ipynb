```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
```

IMPORTING ALL MAJOR LIBRARIES THAT WE WILL USE LATER

```python
sns.set()
#https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
#to know more about iris dataset
data = datasets.load_iris()
print(data["target_names"])
```

problem 1: creating panda dataframe and doins simple stats analysis using graphs

```python
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
print(df.head()) #use to print first five entries
```

```python
print(df.describe()) #use to print the description of dataframe
```

<h1>graph for sepal length variation</h1>

```python
sepal_l = df['sepal length (cm)'].hist()
plt.suptitle("sepal length")
plt.show()
```

<h1>Graph for petal length variation</h1>

```python
petal_l_hist = df['petal length (cm)'].hist()
plt.suptitle("petal length")
plt.show()
```

<h1>Graph for sepal width variation


```python
sepal_w = df['sepal width (cm)'].hist()
plt.suptitle("sepal width")
plt.show()
```

<h1>Graph for petal width variation


```python
petal_w = df['petal width (cm)'].hist()
plt.suptitle("petal width")
plt.show()
```

<h1>Exploratory Data Analysis using pairplot

```python
df['target_name']=df['target'].map({0:"setosa",1:"versicolor",2:"virginica"})
sns.relplot(x="sepal length (cm)", y="target_name",hue ="target_name",data=df)
plt.suptitle("Sepal Length",y=1.1)

```

```python
sns.pairplot(data=df ,diag_kind="hist",hue="target_name")
plt.show()
```
